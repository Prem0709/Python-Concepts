```python
string = 'Asabeneh'
characters = ['A', 'B']
for char in characters:
    if char in string:
        print(f"'{char}' is in '{string}' -> True, position: {string.index(char)}")
    else:
        print(f"'{char}' is in '{string}' -> False")
```
